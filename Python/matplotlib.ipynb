```python
# Simulated data: number of visitors each day for 30 days
days = np.arange(1, 31)
visitors = np.random.randint(1000, 5000, size=30)

# Creating the plot
plt.plot(days, visitors)
plt.title("Website Visitors Over 30 Days")
plt.xlabel("Day")
plt.ylabel("Number of Visitors")
plt.show()
```

```python
# Simulated data: sales amount for 12 months
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = np.random.randint(10000, 50000, size=12)

# Creating the bar chart
plt.bar(months, sales)
plt.title("Sales Performance Over 12 Months")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.show()
```

```python
# Simulated market share data
companies = ['Company A', 'Company B', 'Company C', 'Company D', 'Company E']
market_shares = [35, 25, 15, 10, 15]

# Creating the pie chart
plt.pie(market_shares, labels=companies, autopct='%1.1f%%', startangle=90)
plt.title("Market Share Distribution")
plt.show()
```

```python
# Simulated data for height and weight
heights = np.random.normal(170, 10, 100)  # Heights in cm, mean=170cm, std=10cm
weights = np.random.normal(70, 15, 100)   # Weights in kg, mean=70kg, std=15kg

# Creating the scatter plot
plt.scatter(heights, weights)
plt.title("Scatter Plot of Height vs Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
# Simulated exam scores
scores = np.random.randint(0, 101, size=200)

# Creating the histogram
plt.hist(scores, bins=10, edgecolor='black')
plt.title("Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Number of Students")
plt.show()
```
